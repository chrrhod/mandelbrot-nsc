# Dask Local

### Comments
Running Dask locally we can see that the optimal chunk size is 8 chunks. Which makes sense when comparing it to the amount of performance cores.

we can also see that when using Dask it is roughly 4.6 times slower than using the more manual approach for parallel computing for the standard 1024x1024 size. 

The speed up is comparing to a 1 chunk dask which is run before the runs listed. Something makes it differ each run, this can be seen as the 1 chunk sped up is not 1x.

The serial run is the same gridsize but without dask and no parallel

### 4096 grid size

### Comments
With the increased in grid size to 4096 the optimal chunk size is now 128 chunks.

We can also see the difference between the 1 chunk vs serial is much lower than the previous grid size. This shows that the compute is taking more of the time than the overhead now.
