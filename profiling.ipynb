```python
import cProfile , pstats
from mandelbrot import NaiveMandelbrot , NumpyMandelbrot



xmin, xmax, ymin, ymax = -2.0, 1.0, -1.5, 1.5
width, height = 1024, 1024
max_iter = 100

Naive_version = NaiveMandelbrot(xmin, xmax, ymin, ymax, width, height, max_iter)
        #naive_t, naive_M = benchmark(Naive.compute_mandelbrot)

Numpy_version = NumpyMandelbrot(xmin, xmax, ymin, ymax, width, height, max_iter)


cProfile.run('Naive_version.compute_mandelbrot()','naive_profile.prof')
cProfile.run('Numpy_version.compute_mandelbrot()','numpy_profile.prof')
```

```python
stats = pstats.Stats('naive_profile.prof')
stats.sort_stats('cumulative')
stats.print_stats(10)
```
