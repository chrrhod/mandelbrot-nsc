```python
import cProfile , pstats
from mandelbrot import mandelbrot_naive, mandelbrot_numpy



xmin, xmax, ymin, ymax = -2.0, 1.0, -1.5, 1.5
width, height = 1024, 1024
max_iter = 100


cProfile.run('mandelbrot_naive(xmin, xmax, ymin, ymax, width, height, max_iter)','naive_profile.prof')
cProfile.run('mandelbrot_numpy(xmin, xmax, ymin, ymax, width, height, max_iter)','numpy_profile.prof')
```

```python
stats = pstats.Stats('naive_profile.prof')
stats.sort_stats('cumulative')
stats.print_stats(10)
```

```python

stats = pstats.Stats('numpy_profile.prof')
stats.sort_stats('cumulative')
stats.print_stats(10)
```
